# Diseño Completamente al Azar en Bloques (DBCA)

## Resumen
El **Diseño Completamente al Azar en Bloques (DBCA)** es una variación del DCA donde las unidades experimentales se agrupan en bloques. Los tratamientos se asignan aleatoriamente dentro de cada bloque para controlar la variabilidad de los factores no controlables.

## Modelo Estadístico
El modelo para el DBCA es:

$$
Y_{ij} = \mu + \tau_i + \beta_j + \epsilon_{ij}
$$

Donde:
- $Y_{ij}$: Respuesta observada para el tratamiento $i$ en el bloque $j$.
- $\mu$: Media general.
- $\tau_i$: Efecto del tratamiento $i$.
- $\beta_j$: Efecto del bloque $j$.
- $\epsilon_{ij}$: Error aleatorio.

## Hipótesis
- **$H_0$**: $\tau_1 = \tau_2 = \dots = \tau_t$ (no hay diferencia significativa entre los tratamientos).
- **$H_a$**: Al menos uno de los tratamientos tiene un efecto diferente.

## Supuestos del Modelo

1. **Normalidad de los residuos**: Los residuos deben seguir una distribución normal.
2. **Igualdad de varianzas**: Las varianzas de los tratamientos y bloques deben ser iguales.
3. **Independencia**: Las observaciones deben ser independientes entre sí.

## Tabla ANOVA

La **tabla ANOVA** para el DBCA es:

$$
\begin{array}{|c|c|c|c|c|}
\hline
\textbf{Fuente} & \textbf{GL} & \textbf{Suma de Cuadrados (SC)} & \textbf{Cuadrado Medio (CM)} & \textbf{F} \\
\hline
\text{Tratamientos} & t - 1 & SC_{\text{trat}} = \sum_{i=1}^t \frac{Y_{i.}^2}{r_i} - \frac{Y_{..}^2}{N} & CM_{\text{trat}} = \frac{SC_{\text{trat}}}{t - 1} & F = \frac{CM_{\text{trat}}}{CM_{\text{error}}} \\
\hline
\text{Bloques} & b - 1 & SC_{\text{bloque}} = \sum_{j=1}^b \frac{Y_{.j}^2}{r_j} - \frac{Y_{..}^2}{N} & CM_{\text{bloque}} = \frac{SC_{\text{bloque}}}{b - 1} & F = \frac{CM_{\text{bloque}}}{CM_{\text{error}}} \\
\hline
\text{Error} & (t - 1)(b - 1) & SC_{\text{error}} = SC_{\text{total}} - SC_{\text{trat}} - SC_{\text{bloque}} & CM_{\text{error}} = \frac{SC_{\text{error}}}{GL_{\text{error}}} & \\
\hline
\text{Total} & N - 1 & SC_{\text{total}} = \sum_{i=1}^t \sum_{j=1}^b Y_{ij}^2 - \frac{Y_{..}^2}{N} & & \\
\hline
\end{array}
$$

## Regla de Decisión

- Si $F_{\text{calc}} > F_{\text{tab}}$, entonces **rechazar** $H_0$.
- Equivalente: Si el **p-valor** $< \alpha$, entonces **rechazar** $H_0$.

## Comparaciones Múltiples de Medias

Si el ANOVA indica diferencias significativas, se realizan **pruebas post-hoc** para comparar las medias de los tratamientos. Las pruebas comunes son:

### 1. **Prueba de Diferencia Mínima Significativa (LSD)**

La **LSD** es una prueba post-hoc para realizar comparaciones entre pares de tratamientos. Se utiliza cuando hay una diferencia significativa en el ANOVA.

Fórmula para **LSD**:

$$
LSD = t_{\alpha/2, \, GL_{\text{error}}} \cdot \sqrt{MSE \left( \frac{1}{n_i} + \frac{1}{n_j} \right)}
$$

Donde:
- $t_{\alpha/2, \, GL_{\text{error}}}$: Valor crítico de **t** para el nivel de significancia $\alpha$ y grados de libertad del error.
- $MSE$: Error cuadrático medio del ANOVA.
- $n_i$, $n_j$: Tamaños de muestra de los tratamientos $i$ y $j$.

### 2. **Prueba de Tukey**

La **prueba de Tukey** es una de las pruebas más populares para comparaciones múltiples, especialmente cuando se tienen más de dos tratamientos. La prueba controla el error tipo I a través de la distribución **q** de Studentized Range.

Fórmula para **Tukey**:

$$
HSD = q_{\alpha}(k, \, GL_{\text{error}}) \cdot \sqrt{\frac{MSE}{n}}
$$

Donde:
- $q_{\alpha}(k, \, GL_{\text{error}})$: Valor crítico de la distribución **q** de Studentized Range.
- $MSE$: Error cuadrático medio del ANOVA.
- $n$: Tamaño de la muestra.

### 3. **Prueba de Duncan**

La **prueba de Duncan** es similar a la LSD, pero permite hacer múltiples comparaciones de pares de medias sin ser tan conservadora.

Fórmula para **Duncan**:

$$
HSD = q_{\alpha}(k, \, GL_{\text{error}}) \cdot \sqrt{\frac{MSE}{n}}
$$

### 4. **Prueba de Scheffé**

La **prueba de Scheffé** es una prueba conservadora que realiza comparaciones de medias para cualquier número de tratamientos, incluso cuando las comparaciones no son predefinidas.

Fórmula para **Scheffé**:

$$
S = \sqrt{(k - 1) \cdot F_{\alpha}(k - 1, GL_{\text{error}}) \cdot \frac{MSE}{n}}
$$

Donde:
- $(k - 1)$: Número de comparaciones entre tratamientos.
- $F_{\alpha}(k - 1, GL_{\text{error}})$: Valor crítico de la distribución **F**.
- $MSE$: Error cuadrático medio.
- $n$: Tamaño de la muestra.

### 5. **Prueba de Bonferroni**

La **prueba de Bonferroni** es un método conservador para ajustar el nivel de significancia cuando se realizan múltiples comparaciones.

Fórmula para **Bonferroni**:

$$
\alpha_{\text{adj}} = \frac{\alpha}{k}
$$

Donde:
- $\alpha$: Nivel de significancia original.
- $k$: Número de comparaciones entre medias.

Para cada comparación, se calcula el valor **p** ajustado usando $\alpha_{\text{adj}}$.

---

### Resumen:
- Las pruebas **LSD**, **Tukey**, **Duncan**, **Scheffé** y **Bonferroni** se utilizan para hacer comparaciones entre medias después de realizar un ANOVA significativo.
- Cada prueba tiene su propio nivel de conservadurismo y se elige en función del tipo de diseño experimental y el número de comparaciones a realizar.

ANALISIS DEL EXPERIMENTO 

```python
# Asegurarse de que los detalles del experimento estén definidos
experimento = {
    "factor": "Tratamiento de riego",
    "niveles": ["T1", "T2", "T3", "T4"],  # Los tratamientos en DBCA
    "bloques": ["Bloque I", "Bloque II", "Bloque III"],
    "tratamientos": ["T1", "T2", "T3", "T4"],
    "unidad_experimental": "Parcela",
    "variable_respuesta": "Absorción de agua (kg)"
}

# Acceder a los valores del experimento
factor = experimento['factor']
niveles = experimento['niveles']
bloques = experimento['bloques']
tratamientos_inp = experimento['tratamientos']
unidad_experimental = experimento['unidad_experimental']
variable_respuesta = experimento['variable_respuesta']

# Mostrar resumen del experimento
print("\nResumen del Experimento:")
print(f"Factor: {factor}")
print(f"Niveles del factor: {niveles}")
print(f"Bloques: {bloques}")
print(f"Tratamientos: {tratamientos_inp}")
print(f"Unidad experimental: {unidad_experimental}")
print(f"Variable de respuesta: {variable_respuesta}")
```

Paso 1 : Importar librerías 

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats import diagnostic as diag
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.stattools import durbin_watson
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats import anova


```

Paso 2 : Importar los datos con los que se va a trabajar

```python
# Definir los datos
Rendimiento = [68, 86, 68, 73, 90, 71, 53, 62, 46, 50, 62, 50]
Tratamientos = [1, 1, 1, 2, 2, 2, 3, 3, 3, 4, 4, 4]
Bloque = [1, 2, 3, 1, 2, 3, 1, 2, 3, 1, 2, 3]

# Crear un DataFrame
data = pd.DataFrame({
    'Rendimiento': Rendimiento,
    'Tratamientos': Tratamientos,
    'Bloque': Bloque
})
print(data)
```

Paso 3 : Primer vistazo a la tabla ANOVA

```python
# Modelo ANOVA
model = ols('Rendimiento ~ C(Tratamientos) + C(Bloque)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Mostrar los resultados
print(anova_table)
```

Paso 4 : Extracción de los residuos 

```python
# Obtener los residuos
residuos = model.resid

# Mostrar los residuos
print("Residuos del modelo:")
print(residuos)

# Histograma de los residuos
plt.figure(figsize=(8,6))
plt.hist(residuos, bins=10, edgecolor='black', alpha=0.7)
plt.title('Histograma de los Residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.show()
```

Paso 5 : Graficos de diagnostico

```python
# 4. Gráficos diagnósticos (similar a plot(modelo2) en R)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# a) Residuos vs. Valores ajustados
axes[0, 0].scatter(model.fittedvalues, residuos, color='blue')
axes[0, 0].axhline(0, color='red', linestyle="--")
axes[0, 0].set_xlabel("Valores ajustados")
axes[0, 0].set_ylabel("Residuos")
axes[0, 0].set_title("Residuos vs. Valores ajustados")

# b) Q-Q Plot para evaluar normalidad
sm.qqplot(residuos, line='s', ax=axes[0, 1])
axes[0, 1].set_title("Q-Q Plot")

# c) Gráfico Scale-Location: raíz del valor absoluto de los residuos estandarizados vs. valores ajustados
# Obtener los residuos estandarizados
influence = model.get_influence()
std_resid = influence.resid_studentized_internal
sqrt_abs_std_resid = np.sqrt(np.abs(std_resid))
axes[1, 0].scatter(model.fittedvalues, sqrt_abs_std_resid, color='green')
axes[1, 0].set_xlabel("Valores ajustados")
axes[1, 0].set_ylabel("sqrt(|Residuos estandarizados|)")
axes[1, 0].set_title("Scale-Location")

# d) Residuos vs. Leverage (para detectar puntos influyentes)
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model, ax=axes[1, 1], size=8)
axes[1, 1].set_title("Residuos vs. Leverage")

plt.tight_layout()
plt.show()
```

```python
# Asegurarnos de que Tratamientos sea de tipo categoría
data['Tratamientos'] = data['Tratamientos'].astype('category')

# Diagrama de cajas para visualizar la distribución de los rendimientos por tratamiento
plt.figure(figsize=(6,4))
plt.title("Diagrama de Cajas por Tratamiento")
plt.boxplot([data[data['Tratamientos'] == i]['Rendimiento'] for i in data['Tratamientos'].cat.categories],
            labels=data['Tratamientos'].cat.categories)
plt.xlabel('Tratamientos')
plt.ylabel('Rendimiento')
plt.show()
```

### Interpretación de los Gráficos Diagnósticos

1. **Gráfico: Residuos vs. Valores Ajustados**
   - Este gráfico permite verificar la homogeneidad de varianzas, también conocida como homocedasticidad. En un modelo bien ajustado, los residuos deben estar distribuidos aleatoriamente alrededor de la línea horizontal en el valor 0, sin mostrar patrones. Si los residuos muestran una tendencia (como una curva), esto puede indicar problemas de heterocedasticidad, es decir, que la varianza de los residuos no es constante a lo largo de los valores ajustados.
   
2. **Gráfico: Q-Q Plot de los Residuos**
   - El gráfico Q-Q (cuantil-cuantil) compara la distribución de los residuos con una distribución normal. Si los puntos caen sobre la línea diagonal, indica que los residuos siguen una distribución normal. Cualquier desviación significativa de la línea puede indicar que los residuos no son normales, lo cual podría afectar la validez de los resultados de la ANOVA.
   
3. **Gráfico: Scale-Location (Raíz de los residuos estandarizados vs. Valores Ajustados)**
   - Este gráfico muestra la relación entre los residuos estandarizados y los valores ajustados. Un modelo bien ajustado debería mostrar residuos distribuidos aleatoriamente sin una estructura aparente. Si se observa que los residuos se dispersan de manera desigual, por ejemplo, mostrando más dispersión en los valores más altos o más bajos, esto puede indicar heterocedasticidad. En este caso, se podrían necesitar transformaciones de los datos o considerar modelos alternativos.
   
4. **Gráfico: Residuos vs. Leverage (Detección de Puntos Influyentes)**
   - El gráfico de residuos vs. leverage es útil para identificar puntos de datos influyentes. Los puntos de datos con un alto leverage tienen una mayor influencia en los resultados del modelo, y si estos puntos tienen residuos grandes, podrían estar afectando de manera desproporcionada el ajuste del modelo. Un buen modelo debe tener pocos puntos con un leverage elevado y residuos grandes. Si se encuentran tales puntos, puede ser necesario investigarlos más a fondo.

---

### Interpretación del Diagrama de Cajas

El **diagrama de cajas** muestra la distribución de los rendimientos para cada tratamiento. Se observan las siguientes características clave:

- **La caja**: Representa el rango intercuartílico (IQR), que es el intervalo entre el primer cuartil (Q1) y el tercer cuartil (Q3). La línea en el medio de la caja representa la mediana.
- **Los bigotes**: Se extienden hasta los valores más cercanos a 1.5 veces el rango intercuartílico (IQR) desde los cuartiles. Los puntos fuera de este rango son considerados **outliers** o valores atípicos.
- **Los puntos atípicos**: Si se encuentran puntos fuera de los bigotes, estos valores se consideran atípicos y merecen una revisión adicional para determinar si son errores de medición o si son observaciones válidas que pueden influir en el análisis.

En un **Diseño en Bloques Completamente al Azar (DBCA)**, el diagrama de cajas ayuda a visualizar las diferencias entre los tratamientos. Si los tratamientos muestran una gran variabilidad en el rendimiento o si hay valores atípicos, se deben investigar más a fondo para asegurarse de que no estén afectando los resultados de la ANOVA.


Paso 6 : Verificar los supuestos 

6.1 Prueba de Normalidad kolmogorov-Smirnov

```python
# Prueba de Kolmogorov-Smirnov para la normalidad de los residuos
ks_test = stats.kstest(residuos, 'norm')

# Imprimir el resultado de la prueba
print(f'Prueba de Kolmogorov-Smirnov: Estadístico={ks_test.statistic}, p-valor={ks_test.pvalue}')

# Interpretación del resultado
if ks_test.pvalue > 0.05:
    print("\nInterpretación: El p-valor es mayor que 0.05, por lo tanto, no podemos rechazar la hipótesis nula.\n "
          "Esto sugiere que los residuos siguen una distribución normal y no hay evidencia suficiente para decir lo contrario.")
else:
    print("\nInterpretación: El p-valor es menor que 0.05, lo que indica que podemos rechazar la hipótesis nula.\n"
          "Esto sugiere que los residuos no siguen una distribución normal, lo que podría afectar la validez de los resultados del modelo.")
```

6.2 Prueba de igualdad de varianzas o de homogeneidad de varianzas(Levene)

```python


# Levene's Test for Homogeneity of Variance (center='median')
levene_test_median = stats.levene(
    *[data[data['Tratamientos'] == i]['Rendimiento'] for i in data['Tratamientos'].cat.categories],
    center='median'
)

# Imprimir el resultado de la prueba
print(f'Levene Test (center="median"): Estadístico={levene_test_median.statistic}, p-valor={levene_test_median.pvalue}')

# Interpretación
if levene_test_median.pvalue > 0.05:
    print("\nInterpretación: El p-valor es mayor que 0.05, por lo que no podemos rechazar la hipótesis nula. "
          "Esto sugiere que las varianzas entre los tratamientos  \n son homogéneas y no hay evidencia suficiente para "
          "decir lo contrario.")
else:
    print("\nInterpretación: El p-valor es menor que 0.05, lo que indica que podemos rechazar la hipótesis nula. "
          "Esto sugiere que las varianzas entre los tratamientos  \n no son homogéneas, lo que podría afectar la validez de "
          "los resultados del modelo.")

# Levene's Test for Homogeneity of Variance (center='mean')
levene_test_mean = stats.levene(
    *[data[data['Tratamientos'] == i]['Rendimiento'] for i in data['Tratamientos'].cat.categories],
    center='mean'
)

# Imprimir el resultado de la prueba
print(f'Levene Test (center="mean"): Estadístico={levene_test_mean.statistic}, p-valor={levene_test_mean.pvalue}')

# Interpretación
if levene_test_mean.pvalue > 0.05:
    print("\nInterpretación: El p-valor es mayor que 0.05, por lo que no podemos rechazar la hipótesis nula.\n "
          "Esto sugiere que las varianzas entre los tratamientos  son homogéneas y no hay evidencia suficiente para "
          "decir lo contrario.")
else:
    print("\nInterpretación: El p-valor es menor que 0.05, lo que indica que podemos rechazar la hipótesis nula.\n "
          "Esto sugiere que las varianzas entre los tratamientos no son homogéneas, lo que podría afectar la validez de "
          "los resultados del modelo.")
```

6.3 Prueba de Independencia (Durbin-Watson)

```python

# Realizar el Test de Durbin-Watson sobre los residuos del modelo
dw_test = durbin_watson(residuos)

# Imprimir el resultado del test
print(f'Test de Durbin-Watson: DW = {dw_test}')

# Interpretación
if dw_test < 1.5:
    print("\nInterpretación: El valor de DW es menor que 1.5, lo que indica autocorrelación positiva en los residuos.")
elif dw_test > 2.5:
    print("\nInterpretación: El valor de DW es mayor que 2.5, lo que indica autocorrelación negativa en los residuos.")
else:
    print("\nInterpretación: El valor de DW es cercano a 2, lo que indica que no hay autocorrelación en los residuos.")
```

Paso 7: Confirmacion de tabla ANOVA e interpretación

```python

# Ajustar el modelo ANOVA para DBCA
model = ols('Rendimiento ~ C(Tratamientos) + C(Bloque)', data=data).fit()

# Calcular la tabla ANOVA
anova_table = anova.anova_lm(model, typ=2)

# Mostrar la tabla ANOVA
print("\nTabla ANOVA:")
print(anova_table)

# Interpretación personalizada
print("\nInterpretación de la ANOVA:")

# Verificar si el p-valor de Tratamientos es significativo
p_value_tratamientos = anova_table.loc['C(Tratamientos)', 'PR(>F)']
if p_value_tratamientos < 0.05:
    print(f"El p-valor para el factor '{factor}' es menor que 0.05 \n ({p_value_tratamientos})"
          "Esto indica que hay diferencias significativas en la variable de respuesta entre los tratamientos,\n "
          "es decir, los diferentes tratamientos de riego tienen un impacto significativo en la absorción de agua.")
else:
    print(f"El p-valor para el factor '{factor}' es mayor que 0.05 \n ({p_value_tratamientos}). "
          "Esto sugiere que no hay diferencias significativas entre los tratamientos en cuanto a la absorción de agua.")

# Verificar si el p-valor de Bloque es significativo
p_value_bloques = anova_table.loc['C(Bloque)', 'PR(>F)']
if p_value_bloques < 0.05:
    print(f"El p-valor para el factor '{bloques}' es menor que 0.05 \n ({p_value_bloques}). "
          "Esto indica que los bloques tienen un efecto significativo en la absorción de agua.")
    print("Por lo tanto, se recomienda proceder con comparaciones post-hoc para identificar qué tratamientos \n son significativamente diferentes entre sí.")
else:
    print(f"El p-valor para el factor '{bloques}' es mayor que 0.05 \n ({p_value_bloques}). "
          "Esto sugiere que no hay diferencias significativas entre los bloques en cuanto a la absorción de agua.")
    print("En este caso, no se recomienda proceder con comparaciones post-hoc, ya que no se \n  encontró evidencia suficiente para justificar que los tratamientos difieren significativamente.")

# Resumen adicional de la interpretación
```

Paso 8 : Comparaciones múltiples de medias 

8.1 Prueba de Difrencia Mínima Significativa (LSD)

```python

# Asegurarnos de que la tabla ANOVA esté correctamente generada
anova_table = anova.anova_lm(model, typ=2)

# Obtener el Error Cuadrático Medio (MSE) de la tabla ANOVA
# Usaremos la columna 'sum_sq' (suma de los cuadrados) y 'df' (grados de libertad) para calcular el MSE
MSE = anova_table.loc['Residual', 'sum_sq'] / anova_table.loc['Residual', 'df']

# Calcular la media y la varianza para cada tratamiento
tratamientos = data['Tratamientos'].unique()
medias = data.groupby('Tratamientos')['Rendimiento'].mean()
n = data.groupby('Tratamientos')['Rendimiento'].count()

# Calcular el error estándar de la diferencia de medias (SE_LSD)
SE_LSD = np.sqrt(2 * MSE / n.min())  # Usamos el tamaño mínimo de muestra entre tratamientos

# Calcular el valor LSD
t_critical = stats.t.ppf(1 - 0.025, df=n.sum() - len(tratamientos))  # Valor crítico t
LSD = t_critical * SE_LSD  # Diferencia mínima significativa

# Comparaciones de medias con LSD
print("\nPrueba LSD (Least Significant Difference):")
comparaciones = []

for t1 in tratamientos:
    for t2 in tratamientos:
        if t1 != t2:
            diff = abs(medias[t1] - medias[t2])
            if diff > LSD:
                comparaciones.append((t1, t2, diff, 'Significativa'))
            else:
                comparaciones.append((t1, t2, diff, 'No significativa'))

# Crear DataFrame de comparaciones
comparaciones_df = pd.DataFrame(comparaciones, columns=['Tratamiento 1', 'Tratamiento 2', 'Diferencia', 'Significancia'])

# Mostrar resultados de las comparaciones
print(comparaciones_df)

# Interpretación
print("\nInterpretación de las Comparaciones con LSD:")
for index, row in comparaciones_df.iterrows():
    if row['Significancia'] == 'Significativa':
        print(f"\nComparación significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto indica que las medias de estos tratamientos son significativamente diferentes.")
    else:
        print(f"\nComparación no significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto sugiere que no hay una diferencia estadísticamente significativa entre las medias de estos tratamientos.")
```

8.2 Prueba de Tukey

```python


# Realizar la prueba Tukey HSD para comparaciones múltiples de medias
tukey_result = pairwise_tukeyhsd(data['Rendimiento'], data['Tratamientos'], alpha=0.05)

# Mostrar el resumen de los resultados de Tukey HSD
print("\nResultados de Tukey HSD:")
print(tukey_result)

# Interpretación
print("\nInterpretación de las Comparaciones Múltiples:")

# Recoger los resultados en un DataFrame para una mejor visualización
tukey_df = pd.DataFrame(data=tukey_result.summary().data[1:], columns=tukey_result.summary().data[0])

# Mostrar los resultados de Tukey HSD
print(tukey_df)

# Evaluación e interpretación de los resultados
for index, row in tukey_df.iterrows():
    if row['p-adj'] < 0.05:
        print(f"\nComparación significativa: {row['group1']} vs {row['group2']} (p-valor: {row['p-adj']:.4f})")
        print("Esto indica que las medias de estos tratamientos son significativamente diferentes.")
    else:
        print(f"\nComparación no significativa: {row['group1']} vs {row['group2']} (p-valor: {row['p-adj']:.4f})")
        print("Esto sugiere que no hay una diferencia estadísticamente significativa entre las medias de estos tratamientos.")
```

8.3 Prueba de Duncan

```python

# Realizar la ANOVA
anova_table = anova.anova_lm(model, typ=2)

# Obtener el Error Cuadrático Medio (MSE) de la tabla ANOVA
MSE = anova_table.loc['Residual', 'sum_sq'] / anova_table.loc['Residual', 'df']

# Calcular la media y la varianza para cada tratamiento
tratamientos = data['Tratamientos'].unique()
medias = data.groupby('Tratamientos')['Rendimiento'].mean()
n = data.groupby('Tratamientos')['Rendimiento'].count()

# Número de tratamientos
k = len(tratamientos)

# Duncan's test - Calculando la diferencia mínima significativa
# Calcular el valor crítico de Duncan basado en MSE, número de tratamientos y número de muestras
q_critical = stats.t.ppf(1 - 0.05 / k, df=n.sum() - len(tratamientos))  # Valor t crítico de Duncan
Duncan_LSD = q_critical * np.sqrt(2 * MSE / n.min())  # Diferencia mínima significativa

# Comparaciones de medias con Duncan's test
print("\nPrueba de Duncan (Multiple Range Test):")
comparaciones = []

for t1 in tratamientos:
    for t2 in tratamientos:
        if t1 != t2:
            diff = abs(medias[t1] - medias[t2])
            if diff > Duncan_LSD:
                comparaciones.append((t1, t2, diff, 'Significativa'))
            else:
                comparaciones.append((t1, t2, diff, 'No significativa'))

# Crear DataFrame de comparaciones
comparaciones_df = pd.DataFrame(comparaciones, columns=['Tratamiento 1', 'Tratamiento 2', 'Diferencia', 'Significancia'])

# Mostrar resultados de las comparaciones
print(comparaciones_df)

# Interpretación
print("\nInterpretación de las Comparaciones con Duncan:")
for index, row in comparaciones_df.iterrows():
    if row['Significancia'] == 'Significativa':
        print(f"\nComparación significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto indica que las medias de estos tratamientos son significativamente diferentes.")
    else:
        print(f"\nComparación no significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto sugiere que no hay una diferencia estadísticamente significativa entre las medias de estos tratamientos.")
```

8.4 Prueba de Scheffé

```python

# Realizar la ANOVA
anova_table = anova.anova_lm(model, typ=2)

# Mostrar la tabla ANOVA para ver cómo está estructurada
print("\nTabla ANOVA:")
print(anova_table)

# Obtener el Error Cuadrático Medio (MSE) de la tabla ANOVA
MSE = anova_table.loc['Residual', 'sum_sq'] / anova_table.loc['Residual', 'df']

# Calcular la media para cada tratamiento
tratamientos = data['Tratamientos'].unique()
medias = data.groupby('Tratamientos')['Rendimiento'].mean()
n = data.groupby('Tratamientos')['Rendimiento'].count()

# Número de tratamientos
k = len(tratamientos)

# Grados de libertad
df_1 = k - 1  # Grados de libertad para los tratamientos
df_2 = n.sum() - k  # Grados de libertad para el error

# Valor crítico F de Scheffé
F_critical = stats.f.ppf(1 - 0.05, df_1, df_2)

# Calcular la diferencia mínima significativa (DMS) para Scheffé
DMS = np.sqrt((k - 1) * F_critical * MSE / n.min())  # Usamos el tamaño de muestra mínimo entre tratamientos

# Comparaciones de medias con la prueba de Scheffé
print("\nPrueba de Scheffé (Multiple Range Test):")
comparaciones = []

for t1 in tratamientos:
    for t2 in tratamientos:
        if t1 != t2:
            diff = abs(medias[t1] - medias[t2])
            if diff > DMS:
                comparaciones.append((t1, t2, diff, 'Significativa'))
            else:
                comparaciones.append((t1, t2, diff, 'No significativa'))

# Crear DataFrame de comparaciones
comparaciones_df = pd.DataFrame(comparaciones, columns=['Tratamiento 1', 'Tratamiento 2', 'Diferencia', 'Significancia'])

# Mostrar resultados de las comparaciones
print(comparaciones_df)

# Interpretación
print("\nInterpretación de las Comparaciones con Scheffé:")
for index, row in comparaciones_df.iterrows():
    if row['Significancia'] == 'Significativa':
        print(f"\nComparación significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto indica que las medias de estos tratamientos son significativamente diferentes.")
    else:
        print(f"\nComparación no significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto sugiere que no hay una diferencia estadísticamente significativa entre las medias de estos tratamientos.")
```

8.5 Prueba de Bonferroni

```python

# Realizar la ANOVA
anova_table = anova.anova_lm(model, typ=2)

# Obtener el Error Cuadrático Medio (MSE) de la tabla ANOVA
MSE = anova_table.loc['Residual', 'sum_sq'] / anova_table.loc['Residual', 'df']

# Calcular la media y la varianza para cada tratamiento
tratamientos = data['Tratamientos'].unique()
medias = data.groupby('Tratamientos')['Rendimiento'].mean()
n = data.groupby('Tratamientos')['Rendimiento'].count()

# Número de tratamientos
k = len(tratamientos)

# Comparaciones de medias con LSD (sin Bonferroni)
comparaciones = []
p_valores = []

for t1 in tratamientos:
    for t2 in tratamientos:
        if t1 != t2:
            diff = abs(medias[t1] - medias[t2])
            # Calculamos la diferencia estándar
            SE_LSD = np.sqrt(2 * MSE / n.min())
            t_stat = diff / SE_LSD
            # Calculamos el p-valor sin corrección
            p_valor = 2 * (1 - stats.t.cdf(abs(t_stat), df=n.sum() - k))  # Prueba t bilateral
            comparaciones.append((t1, t2, diff, p_valor))

# Crear DataFrame de comparaciones sin Bonferroni
comparaciones_df = pd.DataFrame(comparaciones, columns=['Tratamiento 1', 'Tratamiento 2', 'Diferencia', 'p-valor'])

# Aplicar corrección de Bonferroni: Multiplicamos los p-valores por el número de comparaciones
comparaciones_df['p-valor ajustado (Bonferroni)'] = comparaciones_df['p-valor'] * len(comparaciones_df)

# Evaluación de la significancia con el p-valor ajustado
comparaciones_df['Significancia'] = comparaciones_df['p-valor ajustado (Bonferroni)'].apply(lambda x: 'Significativa' if x < 0.05 else 'No significativa')

# Mostrar resultados de las comparaciones con Bonferroni ajustado
print("\nResultados de las Comparaciones con Bonferroni:")
print(comparaciones_df)

# Interpretación
print("\nInterpretación de las Comparaciones con Bonferroni:")
for index, row in comparaciones_df.iterrows():
    if row['Significancia'] == 'Significativa':
        print(f"\nComparación significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto indica que las medias de estos tratamientos son significativamente diferentes.")
    else:
        print(f"\nComparación no significativa: {row['Tratamiento 1']} vs {row['Tratamiento 2']} (Diferencia: {row['Diferencia']:.4f})")
        print("Esto sugiere que no hay una diferencia estadísticamente significativa entre las medias de estos tratamientos.")
```
